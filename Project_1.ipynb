```python
# Data visualization
# Example 1: Survival Count

plt.figure(figsize=(6, 4))
sns.countplot(x="Survived", data=data)
plt.title("Survival Count")
plt.show()
```

```python
# Example 2: Age Distribution

plt.figure(figsize=(8, 6))
sns.histplot(data=data, x="Age", hue="Survived", bins=20, kde=True)
plt.title("Age Distribution by Survival")
plt.show()
```

```python
# Example 3: Fare Distribution

plt.figure(figsize=(8, 6))
sns.histplot(data=data, x="Fare", hue="Survived", bins=30, kde=True)
plt.title("Fare Distribution by Survival")
plt.show()
```

```python
# Example 4: Gender vs. Survival

plt.figure(figsize=(6, 4))
sns.countplot(x="Sex", hue="Survived", data=data)
plt.title("Gender vs. Survival")
plt.show()
```

```python
# Example 5: Passenger Class vs. Survival

plt.figure(figsize=(6, 4))
sns.countplot(x="Pclass", hue="Survived", data=data)
plt.title("Passenger Class vs. Survival")
plt.show()
```
